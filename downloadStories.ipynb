```python
import requests,datetime
import urllib,os,sys
import time
import random
import sys,subprocess,re,requests,pickle
import boto3
import requests
import sys

def save_link_to_file(url, folder_mae, folder_time, filename, file_path):
    with open(file_path, 'a') as file:
        file.write(f"{folder_mae},{folder_time},{filename},{url}\n")
    print(f"Link saved to {file_path}")


# Configuração do cliente S3
s3 = boto3.client(
    's3',
    aws_access_key_id='AKIAYNRZ6SIW6AAAQG6U',
    aws_secret_access_key='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',
    region_name='us-east-2'
)
bucket_name = 'nbapedroccm'

def save_to_s3(image_url, folder_mae, folder_time, object_name):
    try:
        response = requests.get(image_url, stream=True)
        if response.status_code == 200:
            s3_file_path = f"{folder_mae}/{folder_time}/{object_name}"
            s3.upload_fileobj(response.raw, bucket_name, s3_file_path)
            print(f"File {s3_file_path} uploaded to S3 bucket {bucket_name}.")
        else:
            print("Failed to get the image.")
    except Exception as e:
        print(f"Error occurred: {e}")
        
def process_and_update_files(pending_file, processed_file):
    while True:
        with open(pending_file, "r") as file:
            lines = file.readlines()

        if not lines:
            break  # Sai do loop se não houver mais linhas para processar

        line = lines.pop(0)  # Remove a primeira linha para processamento

        folder_mae, folder_time, filename, link = line.strip().split(',')
        try:
            save_to_s3(link, folder_mae, folder_time, filename)
            # Adiciona a linha processada ao arquivo de links processados
            with open(processed_file, "a") as proc_file:
                proc_file.write(line)
        except Exception as e:
            print(f"Erro ao processar o link {link}: {e}")
            lines.append(line)  # Adiciona a linha de volta ao final se houver erro

        # Reescreve o arquivo de links pendentes com as linhas restantes
        with open(pending_file, "w") as file:
            file.writelines(lines)

process_and_update_files("all_users_links.txt", "processed_links.txt")
```
