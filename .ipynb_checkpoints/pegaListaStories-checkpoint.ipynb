```python
import boto3
from datetime import datetime

# Configurações do AWS S3
s3_client = boto3.client(
    's3',
    aws_access_key_id='AKIAYNRZ6SIW6AAAQG6U',
    aws_secret_access_key='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',
    region_name='us-east-2'
)

bucket_name = 'nbapedroccm'
prefixes = [
    "bet/",
    "br_a/",
    "br_b/",
    "fash/",
    "fintechs/",
    "food/",
    "foodfash/",
    "laliga/",
    "med/",
    "mkt_digital/",
    "mlb/",
    "nba_times/",
    "nfl_times/",
    "nflplayers/",
    "nhl/",
    "premier_league/",
    "seriea/",
    "soccer_general/",
    "soccer_players/"
]

# Função para encontrar a data do upload mais recente em uma pasta S3
def find_most_recent_upload(bucket, prefix):
    paginator = s3_client.get_paginator('list_objects_v2')
    operation_parameters = {'Bucket': bucket, 'Prefix': prefix}
    page_iterator = paginator.paginate(**operation_parameters)

    most_recent_time = None

    for page in page_iterator:
        if 'Contents' in page:
            for item in page['Contents']:
                if most_recent_time is None or item['LastModified'] > most_recent_time:
                    most_recent_time = item['LastModified']

    return most_recent_time

# Lista para armazenar os resultados
results = []

# Percorrer cada prefixo e encontrar o upload mais recente em suas subpastas
for prefix in prefixes:
    paginator = s3_client.get_paginator('list_objects_v2')
    operation_parameters = {'Bucket': bucket_name, 'Prefix': prefix, 'Delimiter': '/'}
    page_iterator = paginator.paginate(**operation_parameters)

    for page in page_iterator:
        if 'CommonPrefixes' in page:
            for sub_prefix in page['CommonPrefixes']:
                sub_folder = sub_prefix['Prefix']
                most_recent_time = find_most_recent_upload(bucket_name, sub_folder)

                if most_recent_time:
                    results.append((sub_folder, most_recent_time))

# Ordenar os resultados do mais antigo para o mais recente
results.sort(key=lambda x: x[1])

# Exibir os resultados
print("Recentes:")
for result in results:
    folder_name = result[0].strip('/').split('/')[-1]
    date_str = result[1].strftime('%d/%m')
    print(f"{folder_name} - {date_str}")

print("\nANTIGOS:")
for result in results:
    if result[1] < datetime(2024, 6, 20):
        folder_name = result[0].strip('/').split('/')[-1]
        date_str = result[1].strftime('%d/%m')
        print(f"{folder_name} - {date_str}")
```
